```python
from matplotlib import pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline
figure = plt.figure()
ax = Axes3D(figure)
figure.add_axes(ax)
X = np.arange(0,10,0.2)
Y = np.arange(0,10,0.2)
X,Y = np.meshgrid(X,Y)
Z = np.sin(Y+X)
ax.plot_surface(X,Y,Z,rstride=1,cstride=1,cmap='rainbow')
plt.show()
```

```python
from matplotlib import pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
figure = plt.figure()
ax = Axes3D(figure)
figure.add_axes(ax)
X = np.arange(0,10,0.2)
Y = np.arange(0,10,0.2)
X,Y = np.meshgrid(X,Y)
Z = np.cos(X)

ax.plot_surface(X,Y,Z,rstride=1,cstride=1,cmap='rainbow')
plt.show()
```
